### Quick example
The next cell demonstrates a simple recursive function. It is useful for discussing how Python handles function calls and list operations in a compact example.

```
def someGreatFunction(arr):
  if len(arr) <= 1:
    return arr
  pivot = arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle = [x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]
  return someGreatFunction(left) + middle + someGreatFunction(right)

print(someGreatFunction([3,6,8,10,1,2,1]))
```

## Indentation matters in Python
Python uses indentation rather than braces to define code blocks. This is one of the language’s most important style rules.

Code blocks are created with indentation.

- Use the same indentation level consistently within a block.
- A common convention is 4 spaces per indentation level.
- This makes the code easier to read and maintain.

```
def fib(n):
  # Indent level 1: function body
  if n <= 1:
    # Indent level 2: if statement body
    return 1
  else:
    # Indent level 2: else statement body
    return fib(n-1)+fib(n-2)

fib(10)
```

The Fibonacci sequence starts as: 0, 1, 1, 2, 3, 5, ...

## Python is both strongly typed and dynamically typed
This means Python checks types carefully, but you do not need to declare a variable’s type in advance.

- **Strongly typed**: Python does not silently convert values in ways that may surprise you.
- **Dynamically typed**: You can assign a different value type to the same variable later in the program.

This combination makes Python flexible for teaching, experimentation, and data analysis.

## List Slicing

List elements can be accessed in convenient ways.

Basic format: some_list[start_index:end_index]

```
numbers = [0, 1, 2, 3, 4, 5, 6]
```

```
numbers[0:3] == numbers[:3] == [0, 1, 2]
```

```
numbers[5:] == numbers[5:7] == [5, 6]
```

```
numbers[:] == numbers == [0, 1, 2, 3, 4, 5, 6]
```

```
numbers[-1] == 6
```

```
numbers[-3:] == [4, 5, 6]
```

```
numbers[3:-2] == [3, 4]
```

## Object-oriented basics
This section introduces a simple class example to show how Python supports object-oriented programming.

```
class Animal(object):
  def __init__(self, species, age): # Constructor `a = Animal(‘bird’, 10)`
    self.species = species # Refer to instance with `self`
    self.age = age # All instance variables are public

  def isPerson(self): # Invoked with `a.isPerson()`
    return self.species == 'Homo Sapiens'

  def ageOneYear(self):
    self.age += 1

class Dog(Animal): # Inherits Animal’s methods
  def ageOneYear(self): # Override for dog years
    self.age += 7
```

```
a = Animal('bird', 10)
print(a.species, a.age)
```

```
print(a.isPerson())
```
